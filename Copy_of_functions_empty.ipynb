```python
def area_funct(a, b):
  return a*b/2
```

```python
c = hypothenuse_funct(3, 4)
area = area_funct(3, 4)         #area = a*b / 2
altitude = area*2 / c            #area = c*alt / 2
```

```python
def altitude_funct(a,b):
  hyp = hypothenuse_funct(a,b)
  area = area_funct(a,b)
  alt = area*2 / hyp
  return alt
```

## Recursive functions

Define a function to calculate the factorial of a number with recursion!

```python
def fact(n):
  if n==1:
    return 1
  else:
    return n*fact(n-2)
```

```python
fact(5)
```

Calculate the sum of the first n natural numbers!

```python
def rec_sum(n):
  if n==1:
    return 1
  else:
    return n + rec_sum(n-1)
```

```python
rec_sum(10)
```
