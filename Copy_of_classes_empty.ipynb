##METHODS: Functions defined inside of a class

instance methods: take the current instance: "self" as the first argument (just like __init __() )

 --> it will refer to the current instance and perform the function on the current instance!!

```python
class NewCar():
  def __init__(self, brand, model, year, color):
    self.brand = brand
    self.model = model
    self.year = year
    self.color = color
    self.started = False
    self.speed = 0
    self.max_speed = 200

  def start(self):
    print('Starting the car...')
    self.started = True

  def stop(self):
    print('Stopping the car...')
    self.started = False


  def accelerate(self, value):
    if self.started == False:
      print('Car is not started, cannot accelerate!')
      return #we should not go further in the code

    elif(self.speed + value <= self.max_speed):
        self.speed += value
    else: #car is started and increasing the speed is already larger than the max speed
        print('Max speed reached')
        self.speed = self.max_speed
    print('Accelerating to', self.speed, 'km/h')
```

```python
tesla = NewCar('Tesla', 'Y', 2022, 'white')
tesla.start()
```

```python
tesla.accelerate(50)
```

```python
def brake(self, value):
  if self.started == False:
    print('Car is not started, cannot brake!')
    return
  elif (self.speed-value >= 0):
    self.speed -= value
  else: #car is started but reducing the speed would go below 0:
    self.speed = 0
  print('Braking to', self.speed, " km/h")
```

```python
NewCar.brake = brake #Assign funtions to a class -> but "self" has to be the first parameter!
```

```python
mustang = NewCar('Mustang', 'sometype', 2020, "black")
```

```python
#NewCar.start()
NewCar.start(mustang) #--->only this way can I call a function on the CLASS, by providing the exact instance as the 'self' parameter!
```

```python
NewCar.accelerate(mustang, 50) #definition of accelerate: def accelerate(self, value) ! mustang is "self" 50 is the "value"
mustang.accelerate(50)
```

*   calculating the area
*   calculating the circumference






```python
import math
class Circle():
  def __init__(self,radius):
    self.radius=radius
  def area(self):
    return math.pi*self.radius**2
  def circumference(self):
    return 2*math.pi*self.radius
```

Instance-methods:


*   calculates its distance from the (0,0) point
*   calculates the distance from another point
*   calculates the midpoint of the segment defined by itself and another point (argument) --> return a point





```python
import math
```

```python
class Point():
  def __init__(self,x,y):
   self.x=x
   self.y=y
  def distance(self):
    return math.sqrt(self.x**2+self.y**2)
  def distance(self,other):
    return math.sqrt((self.x-other.x)**2+(self.y-other.y)**2)
  def midpoint(self,other):
    return Point((self.x+other.x)/2,(self.y+other.y)/2)
```

```python
#If you want to create a base class-function such that each inherited class has to write its own behavior:

class Person():
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def sayhello(self):
    pass   #"do nothing" - but Python does not let you create a function without a function body
```

```python
class Student(Person):
  def __init__(self, name, age, neptun):
    super().__init__(name,age)
    self.neptun= neptun
  def sayhello(self):
    print('hi')  #overriding the parent class' sayhello function
```
